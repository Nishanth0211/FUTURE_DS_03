```python
# Read the uploaded CSV
df = pd.read_csv("student_feedback.csv")

# Display the first 5Â rows
df.head()
```

```python
# Drop unnecessary columns
df_cleaned = df.drop(columns=['Unnamed: 0', 'Student ID'])

# Check for missing values
df_cleaned.isnull().sum()
```

```python
# Calculate average ratings
avg_ratings = df_cleaned.mean().sort_values(ascending=False)

# Show average ratings
print(avg_ratings)

# Bar plot
plt.figure(figsize=(10,6))
sns.barplot(x=avg_ratings.values, y=avg_ratings.index, palette='coolwarm')
plt.title("Average Student Ratings for Each Feedback Category")
plt.xlabel("Average Rating (1â5 Scale)")
plt.ylabel("Feedback Category")
plt.grid(True)
plt.show()
```
